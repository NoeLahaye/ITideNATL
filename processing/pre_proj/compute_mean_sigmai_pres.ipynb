## compute_mean_sigmai_pres
Compute mean sigmai and pressure from time-averaged temperature and salinity. 
Mean sigmai takes into account grid breathing based on mean SSH (although it is very likely negligible).

Adapted from compute_mean_stratif.ipynb

This notebook is for processing eNATL60 data on occigen

```python
scratch = Path(os.getenv("SCRATCHDIR"))
grid_path = scratch #Path("/store/CT1/hmg2840/lbrodeau/eNATL60/eNATL60-I/")
mean_path = Path("/work/CT1/ige2071/SHARED/mean")

#ssh_fname = "global_mean_gridT-2D.zarr"
temp_fname = "global_mean_gridT.zarr"
salt_fname = "global_mean_gridS.zarr"
#grid_fname = "mesh_mask_eNATL60_3.6.nc"
zgrid_name = "eNATL60_mean_grid_z.zarr" 

out_file = scratch/"global_mean_sigmai.zarr"

#grid_files = [grid_path/grid_fname]
```

```python
### load SSH, temperature and salinity: wrappers
dims = {"ssh": ("x_c", "y_c"), "temp":("x_c", "y_c", "z_c"), "salt":("x_c", "y_c", "z_c")}
dimap = {"x":"x_c", "y":"y_c", "deptht":"z_c"}
inmap = {val:key for key,val in dimap.items()}

def get_var_chkopen(path, varname, chks={}):
    da = xr.open_zarr(path, chunks=chks, overwrite_encoded_chunks=True)[varname].squeeze()
    da = da.rename({dim:dimap[dim] for dim in da.dims})
    #da = da.chunk({dim:chks[dim] for dim in da.dims})
    return da

### the one I should use is this one
def get_var_chkafter(path, varname, chks={}):
    da = xr.open_zarr(path)[varname].squeeze()
    da = da.chunk({dim:chks[dim] for dim in da.dims})
    da = da.rename({dim:dimap[dim] for dim in da.dims})
    return da
```

```python
### load data and arrange coordinates
chunks = {"x_c":200, "y_c":200, "z_c":100, "z_l":100}

ds = xr.open_zarr(scratch/zgrid_name)
if chunks:
    ds = ds.chunk(chunks)
else:
    chunks = ds.chunks
ds = ds.get(["depth_c_3d", "tmask"]).set_coords(["depth_c_3d", "tmask"])

### load and merge temperature and salinity
get_var = get_var_chkafter
chks = {key:ds.chunks[val] for key,val in dimap.items()}
ds["vosaline"] = get_var(mean_path/salt_fname, "vosaline", chks).drop("z_c")
ds["votemper"] = get_var(mean_path/temp_fname, "votemper", chks).drop("z_c")

print("dataset is {:.1f} GB".format(ds.nbytes/1e9))
ds
```

```python
def get_sigmai(ds, return_ds=False, astype=None):
    bvf = eos.sigmai(ds, zcoord="depth_c_3d") .where(ds.tmask)
    if astype:
        bvf = bvf.astype(astype)
    bvf = bvf.reset_coords([co for co in bvf.coords if len(bvf[co].dims)>2], drop=True)
    if return_ds:
        bvf = bvf.to_dataset(name="sigmai")
        for var in bvf.coords:
            bvf[var].encoding.pop("chunks",None) # otherwise writing zarr raise error
        put_attrs = {"grid_file":str(scratch/zgrid_name), "mean_temp":str(mean_path/temp_fname), 
                    "mean_salt":str(mean_path/salt_fname)}
        bvf.attrs = put_attrs
    return bvf
```

```python
from itertools import product
import time

def dist_write_chunks(ds, path, verbose=True, nchk=None):
    """ utilitary function that loops over chunks to store a DataArray in a zarr archive (append mode)"""
    #dim_chk = [next(di for di in da.dims if di.startswith(dim)) for dim in dims]
    if verbose:
        tmei = time.time()
    if isinstance(ds, xr.DataArray):
        name = ds.name if ds.name else "data"
        ds = ds.to_dataset(name=name)
    ### warning this is for testing
    ds = ds.chunk({"z_l":-1})
    
    dims = [*ds.chunks.keys()]
    chks = [np.r_[0, np.array(chk).cumsum()] for chk in ds.chunks.values()]
    if nchk:
        chks = [np.r_[chk[:-1:nchk], chk[-1]] for chk in chks]
        #ds = ds.chunk({di:ds.chunks[di][0]*nchk for di in ["x_c","y_c"] })
    #nks = [len(chk) for chk in ds.chunks.values()]
    nks = [len(chk)-1 for chk in chks] 
    if verbose:
        print("total number of chunks:", np.product(nks), dims, nks)
    ext = [np.arange(nk) for nk in nks]
    
    for chs in product(*ext):
        if verbose:
            tmes = time.time()
        isel = {dim:slice(*chks[ii][chs[ii]:chs[ii]+2]) for ii,dim in enumerate(dims)}
        sds = ds.isel(isel)
        if sds.tmaskutil.any():
            sds.to_zarr(path, mode="a", region=isel)
            if verbose:
                print("chunk {}: {:.0f} s".format(chs, time.time()-tmes), end="; ")
                #print(isel)
        else:
            print("chunk {} is masked: skip".format(chs), end="; ")
    if verbose:
        print("\n finished. Ellapsed time: {:.1f} min".format((time.time()-tmei)/60))
```

```python
sigmai = get_sigmai(ds, return_ds=True, astype="float32")
sigmai
```

```python
### write attribute and create bvf in delayed mode in zarr archive
sigmai.to_zarr(out_file, mode="w", consolidated=True, compute=False)
sigmai.drop_vars("sigmai").to_zarr(out_file, mode="a", compute=True, consolidated=True)
```

```python
%%time
### Compute & store. Distribute over y only
chks_y = np.r_[0, np.array(ds.chunks["y_c"]).cumsum()]
region = {d:slice(0,None) for d in sigmai.dims}
for iy in range(len(chks_y)-1):
    tmes = time.time()
    sliy = slice(chks_y[iy], chks_y[iy+1])
    region["y_c"] = sliy
    sigmai.isel(y_c=sliy).to_zarr(out_file, mode="a", region=region, compute=True)
    print("slice", sliy, time.time()-tmes)
```

## compute mean pressure

```python
### load data and arrange coordinates
chunks = {"x_c":200, "y_c":200, "z_c":100, "z_l":100}
fil_sig = scratch/"global_mean_sigmai.zarr"
out_file = scratch/"global_mean_pres.zarr"
```

```python
ds = xr.open_zarr(scratch/zgrid_name)
if chunks:
    ds = ds.chunk(chunks)
else:
    chunks = ds.chunks
coords = ["e3w", "tmask"]
ds = ds.get(coords).set_coords(coords)
ds_re = xr.open_zarr(fil_sig)
ds_re = ds_re.chunk({d:chunks[d] for d in ds_re.dims if d in chunks})
ds = ds.merge(ds_re)
grid = Grid(ds, periodic=False)
ds
```

```python
pres = var.comp_pres(ds, grid, zmet="e3w", ssh=False).where(ds.tmask).astype("float32")
pres = pres.chunk({"z_c":chunks["z_c"]}).reset_coords("tmask", drop=True)
pres = pres.to_dataset(name="pres")
pres
```

```python
### write attribute and create bvf in delayed mode in zarr archive
pres.to_zarr(out_file, mode="w", consolidated=True, compute=False)
pres.drop_vars("pres").to_zarr(out_file, mode="a", compute=True, consolidated=True)
```

```python
%%time
### Compute & store. Distribute over y only
chks_y = np.r_[0, np.array(ds.chunks["y_c"]).cumsum()]
region = {d:slice(0,None) for d in pres.dims}
for iy in range(len(chks_y)-1):
    tmes = time.time()
    sliy = slice(chks_y[iy], chks_y[iy+1])
    region["y_c"] = sliy
    pres.isel(y_c=sliy).to_zarr(out_file, mode="a", region=region, compute=True)
    print("iy {}: {:.2f} s".format(iy, time.time()-tmes), end=", ")
```
